```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "city": ["Lagos", "Abuja", "Ibadan", "Kano", "PH"],
    "population": [15, 4, 4, 4, 3]  # in millions
}

df = pd.DataFrame(data)
sns.set_theme(style="darkgrid")

plt.figure(figsize=(10, 6))        # 2. chart size
sns.set_palette("dark")

sns.barplot(x='city', y= 'population', data=df,hue='city')
plt.title('city population chart')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

```python
#lagos Market Prices Analysis
# section 1 ( analizing)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('lagos_market_prices.csv')

sns.set_style('darkgrid')
# filter: show only rice row
rice_data = df[df['Item'] == 'Rice']


# filter; show prices above ₦20,000
prices_above = df[df['Price_NGN'] > 20000]

#average price per item
ave_price = df.groupby('Item')['Price_NGN'].mean()#

#Add column — Price_Per_KG = Price_NGN ÷ Quantity_KG
df['Price_Per_KG'] = df['Price_NGN'] / df['Quantity_KG']

print(df.head())
print(df.shape)
print(df.info())
print(df.describe())
print('\nThe Rice Data : ')
print(rice_data)
print('\nThe Prices Above ₦20,000')
print(prices_above)
print('\nThe Average Price Per item:')
print(ave_price)
print('\nPrice per KG:')
print(df[['Item','Price_NGN','Quantity_KG','Price_Per_KG']])








#day 2 charts
#Line chart — tomatoes price over time across markets (hue='Market')

tomatoes_data = df[df['Item'] == 'Tomatoes']

sns.lineplot(x='Year', y='Price_NGN',data=df, hue='Market')

plt.title('Tomatoes price Across Market Chart')
plt.xlabel('prices in NGN')
plt.ylabel('Year')
plt.show()



# Bar chart — average price per food item

ave_per_fooditem = df.groupby('Item')['Price_NGN'].mean()

# Fix: Convert the Series to a DataFrame using reset_index()
ave_per_fooditem_df = ave_per_fooditem.reset_index()

# Fix: Use 'ave_per_fooditem_df' for data, 'Price_NGN' for y-axis, and remove 'hue'
sns.barplot(x='Item', y='Price_NGN', data=ave_per_fooditem_df)


plt.title('average price per food item')
plt.xlabel('Item')
plt.ylabel('Average per Food Item')
plt.show()




# Heatmap — average price by Item vs Market

pivot = df.pivot_table(
    index='Item',  #rows
    columns='Market',# column
    values='Price_NGN',
    aggfunc='mean'   # to get average
)

sns.heatmap(pivot, annot=True, cmap='YlGnBu', fmt='.0f')

plt.title("Average Price per Item across Markets")
plt.show()



#Scatter plot — Price_NGN vs Price_Per_KG


sns.regplot(x='Price_NGN', y= 'Price_Per_KG', data=df)
plt.title(' Prices in NGN vs Prices Per KG Chart')
plt.xlabel('Price_NGN')
plt.ylabel('Price_Per_KG')
plt.show()






#Histogram — distribution of Price_NGN

sns.histplot(x='Price_NGN', kde=True, data=df, binwidth=20)
plt.title('Price in NGN chart')
plt.xlabel('Price in NGN')
plt.ylabel('count')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# NUMPY
# Generate realistic score data
# -----------------------------
np.random.seed(42)

scores = np.random.randint(50, 90, size=100)

# Add abnormal outliers
scores = np.append(scores, [150, 200])

# -----------------------------
# PANDAS
# Structure the dataset
# -----------------------------
df = pd.DataFrame({
    "Student Scores": scores
})

# -----------------------------
# SEABORN
# Statistical visualization
# -----------------------------
sns.set_theme(style="darkgrid")

plt.figure(figsize=(10,6))

sns.boxplot(x=df["Student Scores"])

# -----------------------------
# MATPLOTLIB
# Final presentation controls
# -----------------------------
plt.title("Student Score Distribution")
plt.xlabel("Scores")

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# NUMPY
# Generate fake ML-style features
# -----------------------------
np.random.seed(42)

hours_studied = np.random.randint(1, 10, size=100)

sleep_hours = np.random.randint(4, 9, size=100)

attendance = np.random.randint(50, 100, size=100)

# Simulated exam score relationship
exam_score = (
    hours_studied * 8 +
    sleep_hours * 3 +
    attendance * 0.5 +
    np.random.randint(-10,10,size=100)
)

# -----------------------------
# PANDAS
# Build structured dataset
# -----------------------------
df = pd.DataFrame({
    "Hours Studied": hours_studied,
    "Sleep Hours": sleep_hours,
    "Attendance": attendance,
    "Exam Score": exam_score
})

# -----------------------------
# SEABORN
# Automatic relationship analysis
# -----------------------------
sns.set_theme(style="darkgrid")

sns.pairplot(df)

# -----------------------------
# MATPLOTLIB
# Render final visualization
# -----------------------------
plt.show()
```
